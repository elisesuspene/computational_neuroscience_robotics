# A. Solution numérique des EDPO (méthode d'Euler)

Un système dynamique est un système dont l'évolution dans le temps est décrite par une loi. Il s'agit donc d'un système causal, c’est-à-dire que son avenir ne dépend que de phénomènes du passé ou du présent.

Considérons un système dynamique très simple modélisé par une équation différentielle ordinaire qui décrit l'évolution d'une population de 1000 individus ayant un taux de croissance $k = 0.01$ :

$$\left\{\ \begin{array}{l} \dot{y}(t) = ky(t) \\ y(0) = 1000  \end{array}\right.$$

[La méthode d'Euler](https://fr.wikipedia.org/wiki/M%C3%A9thode_d%27Euler) est une procédure numérique pour résoudre par approximation des équations EDPO avec une condition initiale. 

En utilisant l'approximation de la dérivée $\dot{y}(t) \approx \frac{y(t+1) - y(t)}{dt}$ proposée par Euler, calculer la solution du système via la relation de récurrence :

$$
y(t+1) = y(t) + dt\ \dot{y}(t), t \in [0,n-1] 
$$ 


```python
T = 100.0       # temps de simulation en ans
y_0 = 1000.0    # état initial de la population
dt = 0.08       # pas de temps en ans
k = 0.1         # taux de crouissance

all_y = []      # solution numerique 
times = []      # temps  

t = 0.0
y = y_0
while t < T:
    all_y.append(y)
    times.append(t)
    y_dot = k*y
    y = y + dt*y_dot
    t += dt
```

Comparons maintenant l'approximation d'Euler à la solution analytique du système.

Soit la croissance démographique définie tel que :
$$\frac{\mathrm{d}y(t)}{\mathrm{d}t} = ky(t)$$
Il s'agit d'une équation différentielle du premier ordre séparable qui peut être écrite sous la forme
$$\frac{1}{y(t)}\mathrm{d}y(t) = k\mathrm{d}t$$
En intégrant les deux côtés
$$\int\frac{1}{y(t)}\mathrm{d}y(t) = \int k\mathrm{d}t$$
$$\mathrm{ln}(y(t)) = kt+c$$
$$e^{\mathrm{ln}(y(t))} = e^{kt+c}$$
$$y(t) = e^{kt+c} = Ce^{kt}$$
On évalue en $t=0$ pour obtenir la valeur de $C$ 
$$y(0) = C$$
Ainsi, la solution exacte du système est :
$$y(t) = y(0)e^{kt}$$

```python
sol_exacte = y_0*np.exp(k*np.array(times))

plt.figure(1, figsize=(10,5))
plt.plot(times, all_y, label='Sol. approx.')
plt.plot(times, sol_exacte, label='Sol. exacte')
plt.ylabel("Nombre d'invididus")
plt.xlabel("Temps (ans)")
plt.legend()
```

# B. Modèle de type Intègre-et-tire avec fuite (LIF)

Créer un modèle simple de neurone de type "intègre et tire avec fuite" (LIF = Leaky Integrate and Fire).

Le potentiel membranaire du neurone LIF est donné par :

$$C_\mathrm{m}\cdot\frac{dV}{dt}=I_\mathrm{s} - I_\mathrm{fuite}$$

ou alors

$$C_\mathrm{m}\cdot\frac{dV}{dt}=I_\mathrm{s} - g_\mathrm{fuite}(V-E_\mathrm{fuite})$$


**Objectifs :**
- Créer un modèle de neurone de type LIF
- Faire varier les différents paramètres du modèle
- Identifier leur rôle dans la réponse du neurone
- Analyser la réponse et caractéristiques d'un neurone (courbe IF, raster plot, etc.)

## B.1. Simulation d'un neurone LIF stimulé par un courant constant

### Paramètres

Complétez le code ci-dessous en ajoutant les paramètres du modèles tels que :

$g_{fuite} = 0.385$ mS, $E_{fuite} = -65$ mV, et $C_\mathrm{m} = 3.1$ µF

### Impulsion 

Lorsque $V$ atteint la valeur seuil $V_\mathrm{th} = -50$ mV, le neurone déclenche un potentiel d'action et la tension est réinitialisée à $V_\mathrm{re} = -65$ mV.

### Stimulation

Finalement, appliquez un courant de simulation $I_\mathrm{s} = 6 $ µA durant 1 seconde.


```python
class NeuroneLIF():
    def __init__(self):
        # Paramètres
        self.Cm = 3.1            # capacité membranaire en uF
        self.g_fuite = 0.385       # conductance de fuite en mS
        self.E_fuite = -65.       # potentiel de fuite en mV
        self.Vt = -50.            # valeur seuil d'impulsion
        self.Vre = -65.          # valeur a laquelle la tension se reinitialise
        self.Vm = -65.              # potentiel de la membrane
        self.t = 0.               # instant

        # Variables monitorées
        self.l_spike_t = []       # temps en ms dont les impultions se produissent 
        self.l_Vm = []            # valeurs du potentiel membranaire en mV
        self.l_Is = []            # valeurs de coutants de stimulation en µA
        
    def step(self, dt, Is):
        if self.Vm >= self.Vt :
            self.Vm = self.Vre
            self.l_spike_t.append(self.t)
        else :
            dVm = (Is - self.g_fuite*(self.Vm - self.E_fuite)) / self.Cm
            self.Vm = self.Vm + dt*dVm
        self.t += dt
                    
#Instantiation du neurone LIF
Neurone = NeuroneLIF()

#paramétres de simulation
dt = 0.05  # pas de temps en ms
Is = 6.0  # courant de stimulation en uA
T = 1000.0 # temps de simulation en ms

# Boucle de simulation
times = [i*dt for i in range(int(T/dt))]
for t in times :
    Neurone.l_Vm.append(Neurone.Vm)
    Neurone.l_Is.append(Is)
    Neurone.step(dt, Is)

plt.figure(1, figsize=(10,5))
plt.plot(times, Neurone.l_Vm)
plt.ylabel("Potentiel membranaire (mV)")
plt.xlabel("Temps (ms)")

plt.figure(2, figsize=(10,5))
plt.plot(times, Neurone.l_Is)
plt.ylabel("Courant de stimulation (µA)")
plt.xlabel("Temps (ms)")

plt.show()
```

## B.2. Période réfractaire

Pour plus de réalisme, il est possible d’ajouter au modèle LIF une période réfractaire pendant laquelle le neurone ne pourra pas émettre d'impulsions et la tension de membraine restera à la valeur de reset $V_\mathrm{re}$.

#### Exercises

a) Modifier la classe NeuronLIF afin de permettre la simulation de la période réfractaire.

b) Sous les mêmes conditions de l'exercice précédent, ajouter une période réfractaire de $t_\mathrm{ref} = 20$ ms. 


```python
class NeuroneLIF():
    def __init__(self):
        # Paramètres
        self.Cm = 3.1            # capacité membranaire en uF
        self.g_fuite = 0.385       # conductance de fuite en mS
        self.E_fuite = -65.       # potentiel de fuite en mV
        self.Vt = -50.            # valeur seuil d'impulsion
        self.Vre = -65.          # valeur a laquelle la tension se reinitialise
        self.Vm = -65.              # potentiel de la membrane
        self.t = 0.               # instant
        self.Tref = 20.               # periode refractaire en ms

        # Variables monitorées
        self.l_spike_t = []       # temps en ms dont les impulsions se produisent 
        self.l_Vm = []            # valeurs du potentiel membranaire en mV
        self.l_Is = []            # valeurs de courants de stimulation en µA        
    def step(self, dt, Is):
        last_spike = (0 if len(self.l_spike_t) == 0 else self.l_spike_t[-1])
        if self.t-last_spike < self.Tref :
            self.Vm = self.Vre
        else :
            if self.Vm >= self.Vt :   
                self.Vm = self.Vre
                self.l_spike_t.append(self.t)
            else :
                dVm = (Is - self.g_fuite*(self.Vm - self.E_fuite)) / self.Cm
                self.Vm = self.Vm + dt*dVm
        self.t += dt

#Instantiation du neurone LIF
Neurone = NeuroneLIF()

#paramétres de simulation
dt = 0.05  # pas de temps en ms
Is = 6.0  # courant de stimulation en uA
T = 1000.0 # temps de simulation en ms

# Boucle de simulation
times = [i*dt for i in range(int(T/dt))]
for t in times :
    Neurone.l_Vm.append(Neurone.Vm)
    Neurone.l_Is.append(Is)
    Neurone.step(dt, Is)
    
plt.figure(1, figsize=(10,5))
plt.plot(times, Neurone.l_Vm)
plt.ylabel("Potentiel membranaire (mV)")
plt.xlabel("Temps (ms)")

plt.figure(2, figsize=(10,5))
plt.plot(times, Neurone.l_Is)
plt.ylabel("Courant de stimulation (µA)")
plt.xlabel("Temps (ms)")

plt.show()
```

## B.3. Plusieurs neurones LIF non connectés

### Courbe f(I)

Une courbe f(I) (ou courbe IF) permet de représenter la relation entre la fréquence d'émission des potentiels d'action et l'intensité du courant de stimulation. <p>
La manière la plus simple pour tracer une courve f(I) est de créer plusieurs neurones qui vont recevoir, chacun, une stimulation différente, puis tracer leur fréquence de décharge.
    
Les paramètres et équations sont repris du modèle précédent.
Au moment de créer le modèle des neurones, on change le nombre de neurones (qui correspondra au nombre de points de la courbe), ici N = 300.

Chaque neurone doit recevoir une stimulation différente.
Pour cela, on se sert de la fonction `np.linspace(start, stop, num)` ([documentation](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html)), avec start = min courant, stop = max courant, num = N.

Specifiez une période réfractaire de $t_\mathrm{ref} = 0.5$ ms pour la population.

```python
def simPopulation(T_, N_, Is_, Vm0_):
    #paramétres de simulation
    dt = 0.05  # pas de temps en ms
    T = T_     # temps de simulation en ms
    N = N_     # nb de neurones / nb de points sur la courbe
    
    # création des neurons
    neurons = []
    for n, vm0 in zip(range(N), Vm0_):
        neuron = NeuroneLIF()
        neuron.Tref = 0.5       # periode de refraction
        neuron.Vm = vm0
        neurons.append(neuron)
    
    # Boucle de simulation
    for i in range(N) :
        neuron = neurons[i]
        for j in range(int(T/dt)) :
            neuron.l_Vm.append(Neurone.Vm)
            neuron.l_Is.append(Is[i])
            neuron.step(dt, Is[i])

    return neurons

N = 300
T = 1000.0

# Stimulation
# le 1er neurone reçoit 0µA, ... jusqu'au 300ème neurone qui reçoit 20µA
Is = np.linspace(0.0, 20.0, N)

# état initial 
Vm0 = np.ones((N,))*-65.0 # potentiel membranaire en mV

neurons = simPopulation(T, N, Is, Vm0)
spikes = []
for n in neurons:
    spikes.append(len(n.l_spike_t))
```

Cette fois, il faut tracer la fréquence d'émission des PA (nombre de PA par seconde) en regard du courant de stimulation I.

```python
plt.figure(3, figsize=(8,5))
plt.plot(Is, spikes)
plt.xlabel('Courant de stimulation (µA)')
plt.ylabel("Nombre de PA / seconde (Hz)")
plt.show()
```

### Raster plot

Un *raster plot* est une technique de visualisation de l’activité d'impulsion des neurones au fil du temps. Voici comment tracer un raster plot pour une population de 100 neurones LIF soumis à une stimulation de 10 $\mu$A pendant un temps de simulation de 100 ms :

```python
T = 100 # ms
N = 100 # nb de neurones

# Stimulation
Is = 10*np.ones((N,)) # µA

# état initial 
Vm0 = -60 + np.random.rand(N,1)*10.0 # mV

neurons = simPopulation(T, N, Is, Vm0)
plt.figure(4, figsize=(8,6))

i = 0
for n in neurons:
    neuron_i = np.ones(len(n.l_spike_t))*i
    plt.plot(n.l_spike_t, neuron_i, '.k')
    i += 1

plt.xlabel('Temps (ms)')
plt.ylabel('Index neurone')
plt.show()
```
